# LAB 1 并行快排和串行快排的可视化

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import math
import os
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
def read_performance_file(filename):
    """读取性能数据文件，提取数据规模和耗时"""
    sizes = []
    times = []
    with open(filename, 'r') as f:
        lines = f.readlines()
        for line in lines:
            if '数据规模:' in line and '耗时:' in line:
                parts = line.split(',')
                size = int(parts[0].split(':')[1].strip())
                time = float(parts[1].split(':')[1].split('秒')[0].strip())
                sizes.append(size)
                times.append(time)
    return sizes, times
```

```python
base_path = r"d:\kindoflife\study\Parallel-Algorithm-Analysis-and-Design\LAB 1"

thread2_files = [
    os.path.join(base_path, "parallel_performance_2threads_2cores.txt"),
    os.path.join(base_path, "parallel_performance_2threads_4cores.txt"),
    os.path.join(base_path, "parallel_performance_2threads_8cores.txt"),
    os.path.join(base_path, "parallel_performance_2threads_16cores.txt"),
    os.path.join(base_path, "parallel_performance_2threads_24cores.txt")
]
```

```python
plt.figure(figsize=(12, 8))

for file_path in thread2_files:
    if os.path.exists(file_path):
        sizes, times = read_performance_file(file_path)
        # 从文件名中提取核数
        cores = int(file_path.split("_")[-1].split("cores")[0])
        plt.plot(sizes, times, 'o-', linewidth=2, label=f'{cores}核CPU')

plt.xscale('log')
plt.yscale('log')
plt.xlabel('数据规模')
plt.ylabel('运行时间（秒）')
plt.title('并行快速排序线程数为2时不同CPU核数对比')
plt.grid(True, which="both", ls="-", alpha=0.2)
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
cpu24_files = [
    os.path.join(base_path, "parallel_performance_2threads_24cores.txt"),
    os.path.join(base_path, "parallel_performance_4threads_24cores.txt"),
    os.path.join(base_path, "parallel_performance_8threads_24cores.txt"),
    os.path.join(base_path, "parallel_performance_16threads_24cores.txt"),
    os.path.join(base_path, "parallel_performance_24threads_24cores.txt")
]
```

```python
plt.figure(figsize=(12, 8))

for file_path in cpu24_files:
    if os.path.exists(file_path):
        sizes, times = read_performance_file(file_path)
        threads = int(file_path.split("_")[2].split("threads")[0])
        plt.plot(sizes, times, 's-', linewidth=2, label=f'{threads}线程')

plt.xscale('log')
plt.yscale('log')
plt.xlabel('数据规模')
plt.ylabel('运行时间（秒）')
plt.title('并行快速排序CPU数量为24时不同线程数对比')
plt.grid(True, which="both", ls="-", alpha=0.2)
plt.legend(loc='best')
plt.tight_layout()
plt.savefig(os.path.join(base_path, "cpu24_different_threads.png"), dpi=300)
plt.show()
```

```python
def read_serial_performance(filename):
    """读取串行性能数据文件"""
    serial_sizes = []
    serial_times = []
    with open(filename, 'r') as f:
        lines = f.readlines()
        for line in lines:
            if '数据规模:' in line and '耗时:' in line:
                parts = line.split(',')
                size = int(parts[0].split(':')[1].strip())
                time = float(parts[1].split(':')[1].split('秒')[0].strip())
                serial_sizes.append(size)
                serial_times.append(time)
    return serial_sizes, serial_times
```

```python
serial_file = os.path.join(base_path, "serial_performance.txt")
serial_sizes, serial_times = read_serial_performance(serial_file)
```

```python
plt.figure(figsize=(12, 8))

# 绘制串行性能
plt.plot(serial_sizes, serial_times, 'o-', linewidth=3, color='red', label='串行算法')

# 绘制不同线程数的并行性能
for file_path in cpu24_files:
    if os.path.exists(file_path):
        sizes, times = read_performance_file(file_path)
        threads = int(file_path.split("_")[2].split("threads")[0])
        plt.plot(sizes, times, 's-', linewidth=2, label=f'{threads}线程')

plt.xscale('log')
plt.yscale('log')
plt.xlabel('数据规模')
plt.ylabel('运行时间（秒）')
plt.title('串行与并行快速排序性能对比 (24核CPU)')
plt.grid(True, which="both", ls="-", alpha=0.2)
plt.legend(loc='best')
plt.tight_layout()
plt.savefig(os.path.join(base_path, "serial_vs_parallel.png"), dpi=300)
plt.show()

# 创建加速比图表
plt.figure(figsize=(12, 8))

# 计算并绘制不同线程数的加速比
for file_path in cpu24_files:
    if os.path.exists(file_path):
        sizes, times = read_performance_file(file_path)
        threads = int(file_path.split("_")[2].split("threads")[0])
        
        # 计算加速比
        speedups = []
        for i, size in enumerate(sizes):
            # 找到对应数据规模的串行时间
            if size in serial_sizes:
                serial_idx = serial_sizes.index(size)
                speedup = serial_times[serial_idx] / times[i]
                speedups.append(speedup)
            else:
                # 如果没有对应的串行数据，跳过
                speedups.append(None)
        
        # 过滤掉None值
        valid_sizes = [sizes[i] for i in range(len(sizes)) if speedups[i] is not None]
        valid_speedups = [s for s in speedups if s is not None]
        
        plt.plot(valid_sizes, valid_speedups, 'o-', linewidth=2, label=f'{threads}线程')

# 绘制理想加速比参考线
ideal_x = [min(serial_sizes), max(serial_sizes)]
for thread_num in [2, 4, 8, 16, 24]:
    plt.plot(ideal_x, [thread_num, thread_num], '--', linewidth=1, alpha=0.5, color='gray')
    plt.text(max(serial_sizes), thread_num, f'理想{thread_num}倍加速', va='center', alpha=0.7, fontsize=8)

plt.xscale('log')
plt.xlabel('数据规模')
plt.ylabel('加速比')
plt.title('并行快速排序加速比 (24核CPU)')
plt.grid(True, which="both", ls="-", alpha=0.2)
plt.legend(loc='best')
plt.tight_layout()
plt.savefig(os.path.join(base_path, "speedup_ratio.png"), dpi=300)
plt.show()

# 输出加速比数据表格
print("加速比数据表：")
print("数据规模", end="\t")
for file_path in cpu24_files:
    if os.path.exists(file_path):
        threads = int(file_path.split("_")[2].split("threads")[0])
        print(f"{threads}线程", end="\t")
print()

for i, size in enumerate(serial_sizes):
    print(f"{size}", end="\t")
    for file_path in cpu24_files:
        if os.path.exists(file_path):
            sizes, times = read_performance_file(file_path)
            if size in sizes:
                idx = sizes.index(size)
                speedup = serial_times[i] / times[idx]
                print(f"{speedup:.2f}", end="\t")
            else:
                print("N/A", end="\t")
    print()
```
